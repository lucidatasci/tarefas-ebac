Para aplicao do K-means  necessrio que as variveis sejam numricas. No caso desse dataframe, as variveis no-numricas podem ser facilmente convertidas para variveis numricas, sendo "Month" correspondente ao nmero do ms; "Weekend" e "Revenue" booleanas que podem ser convertidas em 0 e 1.  o que ser feito a seguir.

```python
df["Month"].unique()
```

```python
# Dicionrio de mapeamento de meses
month_mapping = {
    "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "June": 6,
    "Jul": 7, "Aug": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
}

# Substituindo os nomes dos meses pelo nmero correspondente
df['Month'] = df['Month'].map(month_mapping)
```

```python
df["VisitorType"].value_counts()
```

```python
# Alterando a coluna VisitorType para um booleano onde 1  o novo visitante. 

df.rename(columns={'VisitorType': 'NewVisitor'}, inplace=True)

# Excluindo os valores 'Other'
df = df[df['NewVisitor'] != 'Other']

# Definindo o mapeamento diretamente como inteiros
visitor_mapping = {
    'Returning_Visitor': 0,
    'New_Visitor': 1
}

# Aplicando o mapeamento  coluna 'VisitorType' (agora 'NewVisitor')
df['NewVisitor'] = df['NewVisitor'].map(visitor_mapping)
```

```python
df['Revenue'] = df['Revenue'].astype(np.int64) # Convertendo Bool
```

```python
df['Weekend'] = df['Weekend'].astype(np.int64) # Convertendo Bool
```

## Nmero de grupos

Utilize as tcnicas vistas em aula que te ajudem a decidir pelo nmero de grupos a ser utilizados.

```python
#Utilizando o mtodo do cotovelo:

from sklearn.cluster import KMeans
# Variar o nmero de clusters de 1 a 10
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=1111)
    kmeans.fit(df_qt)
    sse.append(kmeans.inertia_)

# Plotar o grfico do mtodo do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Nmero de Clusters')
plt.ylabel('Soma dos Erros Quadrticos (SSE)')
plt.title('Mtodo do Cotovelo')
plt.show()
```

```python
# Utilizando o mtodo da silhueta:

from sklearn.metrics import silhouette_score

# Variar o nmero de clusters de 2 a 10
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=1122)
    labels = kmeans.fit_predict(df_qt)
    silhouette_scores.append(silhouette_score(df_qt, labels))

# Plotar o grfico do mtodo da silhueta
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Nmero de Clusters')
plt.ylabel('Pontuao da Silhueta')
plt.title('Mtodo da Silhueta')
plt.show()
```

Concluso: O mtodo do cotovelo apresenta o nmero de clusters ideal entre 2 ou 3, enquanto o mtodo da silhueta indica 2 como sendo a quantidade ideal. Vamos analisar os 2 casos.

## Avaliao de resultados

Avalie os grupos obtidos com relao s variveis fora do escopo da anlise (minimamente *bounce rate* e *revenue*). 
- Qual grupo possui clientes mais propensos  compra?

```python
df_qt = df.drop(columns=['Revenue'])

# Rodar K-Means com 2 clusters
kmeans_2 = KMeans(n_clusters=2, random_state=42)
df['cluster_2'] = kmeans_2.fit_predict(df_qt)

# Rodar K-Means com 3 clusters
kmeans_3 = KMeans(n_clusters=3, random_state=42)
df['cluster_3'] = kmeans_3.fit_predict(df_qt)

def add_percentage_labels(ax, data, cluster_col): 
    # Calcular as contagens e propores
    counts = data.groupby([cluster_col, 'Revenue']).size().unstack(fill_value=0)
    props = counts.div(counts.sum(axis=1), axis=0)

    # Adicionar rtulos de porcentagem do Revenue=1
    for i in range(len(props)):
        total = counts.iloc[i].sum()
        revenue_1_prop = props.iloc[i].get(1, 0)  # Proporo de Revenue=1
        percentage = f'{revenue_1_prop:.0%}'
        ax.annotate(percentage, (i, revenue_1_prop / 2 + (props.iloc[i, :1].sum() if 1 > 0 else 0)),
                    ha='left', va='bottom', color='black', fontsize=12, weight='regular')

# Grfico de barras empilhadas para clusters com 2 grupos
fig, ax = plt.subplots(figsize=(12, 6))
sns.countplot(x='cluster_2', hue='Revenue', data=df, palette='viridis', ax=ax)
plt.title('Distribuio de Revenue em relao a Clusters (2 Grupos)')
plt.xlabel('Cluster')
plt.ylabel('Contagem')
plt.legend(title='Revenue')
add_percentage_labels(ax, df, 'cluster_2')
plt.show()

# Grfico de barras empilhadas para clusters com 3 grupos
fig, ax = plt.subplots(figsize=(12, 6))
sns.countplot(x='cluster_3', hue='Revenue', data=df, palette='viridis', ax=ax)
plt.title('Distribuio de Revenue em relao a Clusters (3 Grupos)')
plt.xlabel('Cluster')
plt.ylabel('Contagem')
plt.legend(title='Revenue')
add_percentage_labels(ax, df, 'cluster_3')
plt.show()
```

No caso de 2 ou 3 clusters, o segundo cluster possui mais clientes propensos  compra. Respectivamente 30% e 34% dos clientes realizaram a compra, em cada caso.
